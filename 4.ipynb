Jug X capacity = 4 Liters    
Jug Y capacity = 3 Liters    
Initial State = (0,0)     
Goal State = (2,0)    

Allowed Operators:    
1.Fill Jug X    
2.Fill Jug Y     
3.Empty Jug X    
4.Empty Jug Y    
5.Pour X->Y   
6.Pour Y->X   

```python
from collections import deque

x_capacity = int(input("Enter capacity of Jug X: "))
y_capacity = int(input("Enter capacity of Jug Y: "))

goal_x = int(input("Enter desired amount in Jug X: "))
goal_y = int(input("Enter desired amount in Jug Y: "))

start = (0, 0)
goal = (goal_x, goal_y)


def get_next_states(state, x_capacity, y_capacity):
    x, y = state
    states = []

    states.append((x_capacity, y))

    states.append((x, y_capacity))

    states.append((0, y))

    states.append((x, 0))

    transfer = min(x, y_capacity - y)
    states.append((x - transfer, y + transfer))

    transfer = min(y, x_capacity - x)
    states.append((x + transfer, y - transfer))

    return states


def bfs(start, goal, x_capacity, y_capacity):

    visited = set([start])

    queue = deque([[start]])

    while queue:

        path = queue.popleft()

        current = path[-1]

        if current == goal:
            print("Solution Found\n")

            for state in path:
                print(state)

            return

        for neighbour in get_next_states(current, x_capacity, y_capacity):

            if neighbour not in visited:

                visited.add(neighbour)

                queue.append(path + [neighbour])

    print("No solution found.")


bfs(start, goal, x_capacity, y_capacity)
```
