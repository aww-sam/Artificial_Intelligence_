```python
initial_state=(0,0,0,0)
goal_state=(1,1,1,1)
```

```python
def is_valid(state):
    farmer,wolf,goat,cabbage=state

    if wolf==goat and farmer !=wolf:
        return False
    if goat==cabbage and farmer !=goat:
        return False
    return True
```

```python
def get_next_states(state):
    farmer,wolf,goat,cabbage=state

    next_states=[]

    new_state=(1-farmer,wolf,goat,cabbage)

    if is_valid(new_state):
        next_states.append((new_state))
    if farmer==wolf:
        new_state=(1-farmer,1-wolf,goat,cabbage)
    
    if is_valid(new_state):
        next_states.append((new_state))

    if farmer==goat:
        new_state=(1-farmer,wolf,1-goat,cabbage)

    if is_valid(new_state):
        next_states.append((new_state))    

    if farmer==cabbage:
        new_state=(1-farmer,wolf,goat,1-cabbage)

    if is_valid(new_state):
        next_states.append((new_state))   

    return next_states
```

```python
def bfs(initial_state,goal_state):
    
    visited=set([initial_state])

    queue=deque([[initial_state]])

    while queue:

        path=queue.popleft()

        current=path[-1]

        if current==goal_state:
            print("Solution Found!!!\n")

            for state in path:
                print(state)

            return 
        
        for neighbour in get_next_states(current):

            if neighbour not in visited:

                visited.add(neighbour)

                queue.append(path+[neighbour])

    print('No solution Found...')            

bfs(initial_state,goal_state)
```
