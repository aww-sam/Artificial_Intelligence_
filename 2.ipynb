```python
from collections import deque
```

```python
grid=[
    ['S', 'X', '.', 'G', '.'],
    ['.', 'X', 'X', 'X', '.'],
    ['.', 'X', '.', '.', '.'],
    ['.', 'X', '.', 'X', '.'],
    ['.', '.', '.', 'X', '.']
]
rows=5
columns=5
```

```python
def find_positions(grid):
    start=None
    goal=None

    for row in range(rows):
        for col in range(columns):
            if grid[row][col]=='S':
                start=(row,col)
            elif grid[row][col]=='G':
                goal=(row,col)
    return start,goal
```

```python

def is_valid(row,col):
    return(
        0<=row<rows and
        0<=col<columns and
        grid[row][col]!='X'
    )
```

```python
def get_neighbors(state):
    row,col=state

    moves=[
        (row-1,col,'UP'),
        (row+1,col,'DOWN'),
        (row,col-1,'LEFT'),
        (row,col+1,"RIGHT")
    ]

    neighbors=[]

    for R,C,action in moves:
        if is_valid(R,C):
            neighbors.append(((R,C),action))

    return neighbors
```

```python
def robot(initial_state,goal_state):
    visited=set([initial_state])
    queue=deque([[(initial_state,'Start')]])
    while queue:
        path=queue.popleft()
        Current=path[-1][0]
        if Current==goal_state:
            print('Solution FOund!\n')

            for state,action in path:
                print(action)
                print(state)
                print()
            return
        for neighbour,action in get_neighbors(Current):
            if neighbour not in visited:
                visited.add(neighbour)
                new_path = path + [(neighbour, action)]
                queue.append(new_path)
    print("No Solution Found!")
start,goal=find_positions(grid)
robot(start,goal)
```
